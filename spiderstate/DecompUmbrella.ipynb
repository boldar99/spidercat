# Idealised-bus fan-out

A Z spider with `k` outputs: one plain regular edge, plus one X spider per remaining output,
each fanning out into `m_i` idealised edges. Click **Simplify** to collapse every fan into a
bolded idealised bus (and again to expand).

```python
%matplotlib inline
from spiderstate.DecompUmbrella import interactive_fanout

# m1=3, m2=5, m3=2  ->  the Z spider has 4 outputs
diag = interactive_fanout([3, 5, 2])
```

## The full rewrite, side by side

```python
from spiderstate.DecompUmbrella import draw_rewrite

draw_rewrite([3, 5, 2]);
```

## Direct API

`Fanout` is the data model; `z_outputs` is `len(multiplicities) + 1`.

```python
from spiderstate.DecompUmbrella import Fanout

diag = Fanout([3, 5, 2])
print('Z spider outputs:', diag.z_outputs)
diag.draw(title='expanded')
diag.simplify().draw(title='simplified');
```
